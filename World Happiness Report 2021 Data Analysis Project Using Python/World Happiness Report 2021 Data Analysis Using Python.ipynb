```python
#first loading our libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# Setting the parameters that controls the style of the plots eg the color or grid is enable or not

sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
data = pd.read_csv("C:\\Users\\Manish\\world-happiness-report-2021.csv")
```

```python
data.head()
```

```python
#selecting the coloums which we need for our analysis

data_columns = ['Country name', 'Regional indicator', 'Happiness score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption',]
```

```python
#passing new variable name and copying the data
data = data[data_columns].copy()
```

```python
#new variable 'happy_df' and renaming the columns using dictionary
happy_df = data.rename({'Country name': 'country_name', 'Regional indicator':'regional_indicator', 'Happiness score':'happiness_score', 'Logged GDP per capita':'logged_GDP_per_capita', 'Social support':'social_support', 'Healthy life expectancy':'healthy_life_expectancy', 'Freedom to make life choices':'freedom_to_make_life_choices', 'Generosity':'generosity', 'Perceptions of corruption':'perceptions_of_corruption'}, axis =1)
```

```python
#To check columns of our intrests which are been renamed
happy_df.head()
```

```python
#Checking if there are any null values in our dataset
happy_df.isnull().sum()
```

```python
#Creating the plot between happiness score and GDP for different region

plt.rcParams['figure.figsize'] = (15, 7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x = happy_df.happiness_score, y = happy_df.logged_GDP_per_capita, hue = happy_df.regional_indicator, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
```

```python
#creating a pie plot to understand the GDP by region to understand which region has highest % contribution to the world's GPD
#[]has used to sum the values of GDP column
#groupby() function is used to split the data into groups based on criteria as we are finding the % contribution of world's GDP
gdp_region = happy_df.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
gdp_region
```

```python
#now we are going to plot this data into pie plot
# "autopct" enables us to display the percentage value of each slice using Python string formatting.
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

```python
#the backgroud is peach because we have set the parameter "plt.rcParams['figure.facecolor'] = '#FFE5B4'"
#the "Sub-Saharan Africa" is contributing more 20.7% and the lest is "north america and ANZ" 3.1%
```

```python
#finding the total number of countries in each region

total_country = happy_df.groupby('regional_indicator')[['country_name']].count()
print(total_country)
```

```python
#creating a co-relation map to see the relationship that exist between each of the varialbles that are present in dataset
#Corelation Map function('corr')
#Pearson correlation coefficent can be used to summarize the strength of the linear relationship between two data samples
#ax is a matplotlib figures so I have used the subplots fuctions and given figure size
#heatmap function is used in seaborn library, cor=correlation, mask is boolean array or df an optional parameter,
#ata will not be shown in cells if the mask is true and the cells with missing values are automatically masked
#cmap is for colour map, we can customize the color in heatmap ax is where we want to draw the plot

cor = happy_df.corr(method = "pearson")
f, ax = plt.subplots(figsize = (10, 5))
sns.heatmap(cor, mask = np.zeros_like(cor, dtype=np.bool),
            cmap="Blues", square=True, ax=ax)
```

```python
#Reading this corelation matrix

#wherever the cells are blue or dark blue colours this means the variables have the high corelation
#all the cells where there is light blue, greyish or white colour this indicates the variables have very low corelation
#for e.g there is very low or almost negetive corelation between happiness score and perceptions of corruption
#if the citizens of country feel that there is lot of corruption in the country the happiness score would be less
#there is also very low corelation between happiness score and generosity also between healthy life expectancy and perceptio of corruption
#High corelation between happiness score and logged_GDP_per_capita also for social support and healthy life expectancy
```

```python
#Visulization of a bar plot that will display corruption in different regions
# 'groupby' fuction to group all regions
#Finding the average of the corruptions using 'mean' function

corruption = happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

```python
#Central and Eastern Europe has the highest perception of corruption as per the questions answered in the pool
#Visualizing this data using a bar plot
```

```python
#Finding the life expectancy of the top 10 happiest count also the bottom 10 happiest countries
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Perception of Corruption in various Regions')
plt.xlabel('Regions', fontsize = 15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation = 30, ha='right')
plt.bar(corruption.index, corruption.perceptions_of_corruption)
```

```python
top_10 = happy_df.head(10)
bottom_10 = happy_df.tail(10)
```

```python
#ha=horizontal alingment
plt.rcParams['figure.facecolor'] = '#FFE5B4'
fig, axes= plt.subplots(1,2, figsize= (16, 6))
plt.tight_layout(pad= 2)
xlabels= top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x= top_10.country_name, y= top_10.healthy_life_expectancy, ax= axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life expentancy')

xlabels= bottom_10.country_name
axes[1].set_title('Bottom 10 lest happy countries Life Expentancy')
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x= bottom_10.country_name, y= bottom_10.healthy_life_expectancy, ax= axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life expentancy')
```
